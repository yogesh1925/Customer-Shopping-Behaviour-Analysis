```python
High_spenders = df[df['Purchase Amount (USD)']>80]
High_spenders[['Customer ID' ,'Age','Gender','Size','Purchase Amount (USD)']].head(10)
```

```python
Low_spenders = df[df['Purchase Amount (USD)']<30]
Low_spenders[['Customer ID' ,'Age','Gender','Size','Purchase Amount (USD)']].head(10)
```

```python
High_Male_customers = df[(df['Gender'] == 'Male') & (df['Purchase Amount (USD)']>80)]
print('High_purchase_Male_customers:', len(High_Male_customers))
```

```python
High_Female_customers = df[(df['Gender'] == 'Female') & (df['Purchase Amount (USD)']>80)]
print('High_purchase_Female_customers:', len(High_Female_customers))
```

```python
Seasons = ['Winter','Summer']
Seasons_purchase = df[df['Season'].isin(Seasons)]
print('selected seasonal purchase:',len(Seasons_purchase))
```

```python
Missing_ratings = df[df['Review Rating'].isnull()]
print(len(Missing_ratings)) #finding null
```

```python
avg_review_rating = df['Review Rating'].mean()
df['Review Rating'] = df['Review Rating'].fillna(avg_review_rating)
print(df[df['Review Rating'] == avg_review_rating])
```

```python
Missing_ratings = df[df['Review Rating'].isnull()]
print(len(Missing_ratings)) #null values in review ratings are replaced by avg review rating
```

```python
print(df.duplicated().sum())

df.drop_duplicates() #if present
```

```python
df = df.rename(columns = {'Purchase Amount (USD)' : 'Purchase Amount'})
print(df.columns.tolist())
```

```python
def age_group (age):
    if age < 25 :
        return 'younger (less than 25)'
    elif age < 45:
        return 'Middle (<45)'
    else:
        return 'Senior(45 above)'
df['Age_group'] = df['Age'].apply(age_group)
df[['Age','Age_group']].head(5)
```

```python
df['Spend levels'] = df['Purchase Amount'].apply(lambda x : 'High' if x >= 80 else 'Low' ) 
df[['Purchase Amount','Spend levels','Gender']].head(5)
```

```python
df['Gender_upper'] = df['Gender'].str.upper()
df[['Gender_upper']].head(5)
```

```python
df_sorted = df.sort_values('Purchase Amount',ascending = False)
df_sorted[['Customer ID','Purchase Amount']].head(5) #using sort
```

```python
df['Rank'] = df['Purchase Amount'].rank(method = 'dense',ascending=False)
df[df['Rank'] == 1][['Customer ID','Rank']].head(5) #using dense rank
```

```python
df['Season'].value_counts()
```

```python
df['Category'].unique() #array
df['Category'].unique().tolist() #list
```

```python
df.groupby('Category')['Purchase Amount'].sum().sort_values(ascending = False)
```

```python
category_summary = df.groupby('Category')['Purchase Amount'].agg([
    ('total sales','sum'),
    ('average spend','mean')
]).round(2)

category_summary
```

```python
pivot = df.pivot_table(
    values='Purchase Amount',
    index='Season',
    columns='Gender',
    aggfunc='mean'
).round(2)

pivot #excel like table
```

```python
pd.crosstab(df['Gender'],df['Subscription Status'])
```

```python
df_purchase = df[['Customer ID', 'Purchase Amount', 'Category']].head(5)
df_ratings = df[['Customer ID', 'Review Rating', 'Gender']].head(5)

inner_merge = pd.merge(df_purchase,df_ratings, on = 'Customer ID', how = 'inner' )
inner_merge 
```
